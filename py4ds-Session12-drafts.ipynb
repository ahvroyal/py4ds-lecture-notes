```python
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312, 'Illinois': 149995})

pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860, 'Illinois': 12882135})
```

```python
df = pd.DataFrame({'area':area, 'pop':pop})
df
```

```python
# accessing a dataframe column
```

```python
# dict-like style
df['area']
```

```python
df['pop']
```

```python
# attribute-like style (not recommended)
df.area
```

```python
df['area'] = df['area']*1000000
df
```

```python
df['density'] = df['pop'] / df['area']
df
```

```python
df[['pop', 'density']]
```
