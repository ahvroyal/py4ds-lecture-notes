### Masks

```python
import pandas as pd
```

```python
df = pd.read_csv('president_heights.csv')
df.head()
```

```python
h = df['height(cm)'].values
h
```

```python
x = h > 180
x
```

```python
h[x]
```

```python
h[h>180]
```

```python
h[(h > 180) & (h < 185)]
```
