### regular expressions

```python
df['name'].str.match('^R[a-zA-Z]*$')
```

```python
df
```

```python
df['name'].str.match('^[A-Z][a-z]*$')
```

```python
df = pd.DataFrame({'age': [25, 33, 14, 51],
                   'test': ['My name is: Reza', 'My name is: Kourosh',
                            'My name is: Sahar', 'My name is: Rahman',]})
df
```

```python
df['test'].str.extract('^My name is: ([A-Z][a-z]*)$')
```

```python
df['test'].str[12:] # you may say we can easily do this, but it's not always that easy
```

```python
# for example :
df = pd.DataFrame({'age': [25, 33, 14, 51],
                   'test': ['My name IS: Reza', 'My beautiful name is: Kourosh',
                            'My name : Sahar', 'My name is: Rahman',]})
df
```

```python
df['test'].str[12:] # oops !
```

```python
# so it's wise to use a regular expression
```

```python
df['test'].str.extract(': ([A-Z][a-z]*)')
```
